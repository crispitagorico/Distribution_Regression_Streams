```python
kw = dict(marker='o', linestyle='none', color='r',alpha=0.8,legend=False)
from scipy import stats

X = [e.copy() for e in X_disaggr]
indices = []

new_list = []
clim_col = ['temperature','moisture']
for f in trange(len(X)):
    
    pix_col = []
    for col in X[f].columns:
        if 'pix' in col:
            pix_col.append(col)
    
    # remove rows with all nans
    X_new = X[f][pix_col].dropna(how='all',axis=0)

    # linearly interpolate
    X_new = X_new.interpolate(method='linear')
    X_new = X_new.fillna(method='ffill').fillna(method='bfill')
    
   
    # remove all pixels for which the time series deviates too much from the "mean" 
    
    #X[f].index = np.arange(len(X[f]))
    #X[f]['time']=X[f].index/len(X[f])
    

    #fig, ax = plt.subplots(1,3,figsize=(20,5))
    #X_new.plot(legend=False,ax=ax[0])

    X_new = X_new.T[(np.abs(stats.zscore(X_new.T)) < 3).all(axis=1)]
    X_new = X_new.T
    
    #X_new.plot(ax=ax[1],legend=False)
    
    # remove all timestamps which correspond to outlier
    #X_new = X_new.ewm(alpha=0.6).mean()
    
    X_filter = X_new.rolling(window=5, center=True,min_periods=0).median()
    X_filter.iloc[-3:,:]=X_new.iloc[-3:,:]
    diff = np.abs(X_filter-X_new)
    outlier = diff>0.1
    
    X_new[outlier]=np.nan
    X_new = X_new.interpolate(method='linear')
    X_new = X_new.fillna(method='ffill').fillna(method='bfill')

    new_df =X[f][clim_col].join(X_new)
    #new_df = new_df.interpolate(method='linear')
    #new_df = new_df.fillna(method='ffill').fillna(method='bfill')
    
    
    
    #new_df[list(X_new.columns[:5])].plot(legend=False)
    #break
    
#     #X_new.plot(ax=ax[2],legend=False)
#     #plt.show()
    
#     new_list.append(X_new)
#     indices.append(X_new.index.values)
    


    
    
```
